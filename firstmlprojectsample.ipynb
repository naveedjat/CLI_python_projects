**Step 1: Import Libraries**


```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error

```

**Step 2 : Create Simple dataset**


```python
data = {
    'size':[1000,1500,2000,2500,3000],
    'price':[200000,300000,400000,500000,600000]
}

df = pd.DataFrame(data)
df
```

**Step 3: Define X (input) and y (output)**

```python
X = df[['size']] # feature
y = df['price'] # target
```

**Step 4: Split Data (VERY IMPORTANT)**

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

**Step 5: Choose Model**

```python
model = LinearRegression()
```

**Step 6: Train Model**

```python
model.fit(X_train,y_train)
```

**Step 7: Make Predictions**

```python
predictions = model.predict(X_test)
```

**Step 8: Evaluate Model**

```python
error = mean_absolute_error(y_test, predictions)
print("Error:", error)
```

**Step 9: Predict New Value (Real Use)**

```python
# model.predict([[2200]])  this works well but it is getting error or warning which should not be in production so we use different method to minize this warning.
```

```python
new_data = pd.DataFrame({"size": [3500]})
model.predict(new_data)
```
